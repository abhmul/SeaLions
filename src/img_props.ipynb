```python
from keras.preprocessing.image import load_img, img_to_array
import numpy as np
import matplotlib.pyplot as plt

img = load_img("../Train/41.jpg")
imarr = img_to_array(img)

plt.imshow(imarr * 1./255)
plt.show()
```

```python
plt.hist(imarr.flatten())
plt.show()
```

```python
print(imarr.shape)
for row in imarr.transpose(2, 0, 1).reshape(3, -1):
    plt.hist(row)
    plt.show()
```

```python
plt.imshow(imarr[2900:3100, 2900:3100] * 1./255)
plt.show()

plt.imshow(imarr[2920:2930, 2915:2925] * 1./255)
plt.show()

diff = np.abs(imarr - imarr[2923, 2914])
thresh = ((1 - (diff * 1./255)) > .9).astype(np.float32)
thresh = np.average(thresh, axis=2)
plt.imshow(thresh, cmap='gray')
plt.show()

plt.imshow(imarr * 1./255)
plt.show()

plt.imshow(thresh[:, :, np.newaxis] * imarr)
plt.show()
```
